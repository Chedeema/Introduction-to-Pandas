```python
import pandas as pd

# Load the dataset
df = pd.read_csv('employee_data.csv')

# Display first five rows
print(df.head())

# Check for missing values
print(df.isna().sum())

# Fill missing values in Bonus column with the median bonus
df['Bonus'].fillna(df['Bonus'].median(), inplace=True)

# Ensure no missing values in EmployeeID, Name, or Salary columns
df.dropna(subset=['EmployeeID', 'Name', 'Salary'], inplace=True)

# Check the result
print(df.isna().sum())
Output:


# Filter employees with salary greater than 100,000
high_earners = df[df['Salary'] > 100000][['EmployeeID', 'Name', 'Salary', 'Department']]
print(high_earners)




# Group by department and calculate mean salary and sum of bonus
dept_agg = df.groupby('Department').agg({'Salary': 'mean', 'Bonus': 'sum'})
print(dept_agg)
Output:


# Convert HireDate to datetime and sort by it
df['HireDate'] = pd.to_datetime(df['HireDate'])
sorted_df = df.sort_values(by='HireDate')
print(sorted_df[['EmployeeID', 'Name', 'HireDate']])
Output:

yaml
Copy code
 EmployeeID          Name   HireDate
7          108    Hank Green 2015-08-25
6          107    Gina White 2016-02-05
3          104    David Clark 2016-10-01
1          102    Bob Johnson 2017-01-15
9          110    Jack Wilson 2018-06-22
Save Results

python
Copy code
# Save high earners and department aggregation to CSV files
high_earners.to_csv('high_earners.csv', index=False)
dept_agg.to_csv('department_aggregation.csv', index=False)
```
